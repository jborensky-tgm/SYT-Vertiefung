```r
start_time <- Sys.time()

library(ggplot2)
library(reshape2)
library(Amelia)

path = "train.csv"
train=read.csv(path)
knitr::kable(head(train), format = "markdown")
#train = as.matrix(train)

missmap(train, col=c("black", "grey"))
# plot(Sex ~ Survived, data=train)
# ggplot(train, aes(x = Survived)) +
#  geom_bar(width=0.5, fill = "coral") +
#  geom_text(stat='count', aes(label=stat(count)), vjust=-0.5) +
#  theme_classic()
ggplot(train, aes(x = Survived, fill=Sex)) +
 geom_bar(position = position_dodge()) +
 geom_text(stat="count", 
           aes(label=stat(count)), 
           position = position_dodge(width=1), vjust=-0.5)+
 theme_classic()

# heatmap(as.matrix(train))

# hist(train[,'Pclass'],xlab = "Pclass",col = "yellow",border = "blue")

# ggplot(train, aes(x=train[,'cyl'], y=train[,'hp'])) + geom_point()

cormat <- round(cor(train[sapply(train, is.numeric)]),2)
melted_cormat <- melt(cormat)
ggplot(data = melted_cormat, aes(x=Var1, y=Var2, fill=value)) + 
  geom_tile()
```

```r
na_count <- data.frame(sapply(train, function(y) sum(length(which(is.na(y))))))
knitr::kable(na_count, format = "markdown")
```
